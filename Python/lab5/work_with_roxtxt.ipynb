### Finding the name by number before it

```python
import re

path = r'C:\PP2Main\Python\lab5\tasks\row.txt'

with open(path, 'r', encoding='utf-8') as file:
    data = file.read()

# Нахождение товара по цифре и вводы названия.

pattern = r"(\d+)\.\s*([A-Za-zА-Яа-я\[]+.+)" 
"""
(\d+) - группа чисел
\. - точка после числа
\s* - так как в файле пробелы и вынос строки, ищем их
(.+) - само название продукта до выноса
([A-Za-zА-Яа-я\[]+.+ - проверка на то, что следующее будет буква, скобка и т.д чтобы исключить даты в конце документа
"""
products = re.findall(pattern, data)


for product in products:
    print(product)
```

### Finding the product and price of it

```python
import re

path = r'C:\PP2Main\Python\lab5\tasks\row.txt'

with open(path, 'r', encoding='utf-8') as file:
    data = file.read()

# Нахождение товара по цифре и вводы названия.

pattern = r"(\d+)\.\s*([A-Za-zА-Яа-я\[]+.+)"
products = re.findall(pattern, data)


pattern_price = r"Стоимость\s+([\d\s]+,\d{2})"
"""
\s+ - ищем после "Стоимость" пробелы
([\d\s]+,\d{2}) - группа, где находятся целая часть, после нее пробел, и вещественная(состоящяя из 2 цифр)
"""
prices = re.findall(pattern_price, data)


for product, price in zip(products, prices):
    print(f"{product}, Цена: {price}")
```
