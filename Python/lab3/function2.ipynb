##### 1. Write a function that takes a single movie and returns `True` if its IMDB score is above 5.5

##### 3. Write a function that takes a category name and returns just those movies under that category.

```python
# Dictionary of movies

movies = [
{
"name": "Usual Suspects", 
"imdb": 7.0,
"category": "Thriller"
},
{
"name": "Hitman",
"imdb": 6.3,
"category": "Action"
},
{
"name": "Dark Knight",
"imdb": 9.0,
"category": "Adventure"
},
{
"name": "The Help",
"imdb": 8.0,
"category": "Drama"
},
{
"name": "The Choice",
"imdb": 6.2,
"category": "Romance"
},
{
"name": "Colonia",
"imdb": 7.4,
"category": "Romance"
},
{
"name": "Love",
"imdb": 6.0,
"category": "Romance"
},
{
"name": "Bride Wars",
"imdb": 5.4,
"category": "Romance"
},
{
"name": "AlphaJet",
"imdb": 3.2,
"category": "War"
},
{
"name": "Ringing Crime",
"imdb": 4.0,
"category": "Crime"
},
{
"name": "Joking muck",
"imdb": 7.2,
"category": "Comedy"
},
{
"name": "What is the name",
"imdb": 9.2,
"category": "Suspense"
},
{
"name": "Detective",
"imdb": 7.0,
"category": "Suspense"
},
{
"name": "Exam",
"imdb": 4.2,
"category": "Thriller"
},
{
"name": "We Two",
"imdb": 7.2,
"category": "Romance"
}
]
def category_in_dict(category_name):
    for movie in movies:
        if movie["category"] == category_name:
            return True
    return False

def movies_in_category(category_name):
    movies_in_category = []
    for movie in movies:
        if movie["category"] == category_name:
            movies_in_category.append(movie["name"])
    return movies_in_category


# Example usage
category_name = input("Enter the category of the movie: ")

if not category_in_dict(category_name):
    print(f"{category_name} is not in the dictionary")

else:
    print(f"Movies in category -- {category_name}", end="\n")
    for movie in movies_in_category(category_name):
        print(movie, end=" | ")
```

##### 4. Write a function that takes a list of movies and computes the average IMDB score.

```python
# Dictionary of movies

movies = [
{
"name": "Usual Suspects", 
"imdb": 7.0,
"category": "Thriller"
},
{
"name": "Hitman",
"imdb": 6.3,
"category": "Action"
},
{
"name": "Dark Knight",
"imdb": 9.0,
"category": "Adventure"
},
{
"name": "The Help",
"imdb": 8.0,
"category": "Drama"
},
{
"name": "The Choice",
"imdb": 6.2,
"category": "Romance"
},
{
"name": "Colonia",
"imdb": 7.4,
"category": "Romance"
},
{
"name": "Love",
"imdb": 6.0,
"category": "Romance"
},
{
"name": "Bride Wars",
"imdb": 5.4,
"category": "Romance"
},
{
"name": "AlphaJet",
"imdb": 3.2,
"category": "War"
},
{
"name": "Ringing Crime",
"imdb": 4.0,
"category": "Crime"
},
{
"name": "Joking muck",
"imdb": 7.2,
"category": "Comedy"
},
{
"name": "What is the name",
"imdb": 9.2,
"category": "Suspense"
},
{
"name": "Detective",
"imdb": 7.0,
"category": "Suspense"
},
{
"name": "Exam",
"imdb": 4.2,
"category": "Thriller"
},
{
"name": "We Two",
"imdb": 7.2,
"category": "Romance"
}
]
def average_imdb_score(movie_list):
    total_imdb_score = 0
    for movie in movies:
        if movie["name"] in movie_list:
            total_imdb_score += movie["imdb"]
    return total_imdb_score / len(movie_list)

# Example usage

n = int(input("How many movies to display: "))
movie_list = []
for i in range(n):
    movie_name = input(f"Enter movie name {i+1}: ")
    movie_list.append(movie_name)

if average_imdb_score(movie_list) == 0:
    print("No movies found")
else:
    print(f"Average IMDB score of the movies (which was find in dictionary) is {average_imdb_score(movie_list)}")
```

#### 5. Write a function that takes a category and computes the average IMDB score.

```python
movies = [
{
"name": "Usual Suspects", 
"imdb": 7.0,
"category": "Thriller"
},
{
"name": "Hitman",
"imdb": 6.3,
"category": "Action"
},
{
"name": "Dark Knight",
"imdb": 9.0,
"category": "Adventure"
},
{
"name": "The Help",
"imdb": 8.0,
"category": "Drama"
},
{
"name": "The Choice",
"imdb": 6.2,
"category": "Romance"
},
{
"name": "Colonia",
"imdb": 7.4,
"category": "Romance"
},
{
"name": "Love",
"imdb": 6.0,
"category": "Romance"
},
{
"name": "Bride Wars",
"imdb": 5.4,
"category": "Romance"
},
{
"name": "AlphaJet",
"imdb": 3.2,
"category": "War"
},
{
"name": "Ringing Crime",
"imdb": 4.0,
"category": "Crime"
},
{
"name": "Joking muck",
"imdb": 7.2,
"category": "Comedy"
},
{
"name": "What is the name",
"imdb": 9.2,
"category": "Suspense"
},
{
"name": "Detective",
"imdb": 7.0,
"category": "Suspense"
},
{
"name": "Exam",
"imdb": 4.2,
"category": "Thriller"
},
{
"name": "We Two",
"imdb": 7.2,
"category": "Romance"
}
]

def average_imdb_score(category):
    total_imdb_score = 0
    count = 0
    for movie in movies:
        if movie["category"] == category:
            total_imdb_score += movie["imdb"]
            count += 1
    if count == 0:
        return 0
    return total_imdb_score / count

# Example usage

category = input("Enter the category of the movie: ")
if average_imdb_score(category) > 0:
    print(f"Averege IMDB score of movies in {category} category is {average_imdb_score(category)}")
else:
    print(f"No movies found in {category} category")

```
